### API calls from RAPID API 
Alpha Vantage API to collect stock market data

```python
%pip install requests
```

```python
import requests

company = "State Bank of India"

url = "https://alpha-vantage.p.rapidapi.com/query"

querystring = {"datatype":"json","keywords":company,"function":"SYMBOL_SEARCH"}

headers = {
	"x-rapidapi-key": "<API-KEY-HERE>",
	"x-rapidapi-host": "alpha-vantage.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

response.json()
```

### Getting the data for State Bank of India from above symbol is SBIN.BSE

```python
symbol = "SBIN.BSE"

url = "https://alpha-vantage.p.rapidapi.com/query"

querystring = {"function":"TIME_SERIES_DAILY","symbol":symbol,"outputsize":"compact","datatype":"json"}

headers = {
	"x-rapidapi-key": "<API-KEY-HERE>",
	"x-rapidapi-host": "alpha-vantage.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

data = response.json()
```

```python
data
```

```python
data.keys()
```

```python
data['Time Series (Daily)']
```

```python
import pandas as pd
df = pd.DataFrame(data['Time Series (Daily)']).T
df.head()
```

```python
df.info()
```

```python
df.index
```

### Convert index to datetime

```python
df.index = pd.to_datetime(df.index)
```

```python
df.index
```

### convert all data into float

```python
for i in df.columns:
    df[i] = df[i].astype(float)
```

```python
df
```

```python
df.index.name = "Date"
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.to_csv("SBIStockData.csv")
```

### Candlestick Charts

```python
%pip install plotly nbformat
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(
    data = [
        go.Candlestick(
            x = df.index,
            open = df['1. open'],
            high = df['2. high'],
            low = df['3. low'],
            close = df['4. close']
        )
    ]
)
fig.update_layout(
    width=800,
    height=600
)
fig.show()
```
