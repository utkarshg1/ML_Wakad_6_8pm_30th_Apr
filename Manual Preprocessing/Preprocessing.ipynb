### Step 4 - Replace the missing values

### Function called as df.fillna(value)

```python
def replacer(df):
    cat = list(df.columns[df.dtypes=='object'])
    for i in df.columns:
        if i in cat:
            md = df[i].mode()[0]
            df[i] = df[i].fillna(md)
        else:
            mn = df[i].mean()
            df[i] = df[i].fillna(mn)
    print('Missing values replaced')
```

```python
replacer(df)
```

### Step 5 : Seperate X and Y(Weight)

### Step 6: Perform preprocessing on X

```python
X_cat = X.select_dtypes(include='object')
X_cat
```

```python
df['Cylinders'].unique()
```

```python
df['Cylinders'].value_counts()
```

```python
X_con = X.select_dtypes(exclude='object')
X_con
```

### Apply one hot ecoding on categorical features

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore',
                    sparse_output=False).set_output(transform='pandas')
X_cat_pre = ohe.fit_transform(X_cat)
X_cat_pre.head()
```

```python
X_cat_pre.to_csv("ohe.csv", index=False)
```

### Apply standard scaler on numerical features

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().set_output(transform='pandas')
X_con_pre = scaler.fit_transform(X_con)
X_con_pre.head()
```

```python
X_con_pre.shape
```

### Combine both categorical and continous preprocessed data

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

```python
X_pre.to_csv("pre.csv", index=False)
```

### Apply Train Test split
1. 80% of data will be Train : Will always be passed to model
2. 20% of data will be Test : These values are hidden from model



### Random state acts like seed for random 80%, 20% split , for a given random state split will always be same
Test data split should be between 20% to 30%

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

### Preprocessing is completed

### Build the model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

### Evaluate the model

```python
# Get r2 score for train
model.score(xtrain, ytrain)
```

```python
# Get r2 score for test
model.score(xtest, ytest)
```

### Predict the result for train and test

```python
ypred_train = model.predict(xtrain)
ypred_test = model.predict(xtest)
```
